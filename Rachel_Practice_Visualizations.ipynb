## 1. Matplotlib
Matplotlib is a versatile library for creating static, animated, and interactive plots.

```python
# Import Matplotlib
import matplotlib.pyplot as plt
import numpy as np

# Generate sample data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create a basic line plot
plt.figure(figsize=(8, 5))
plt.plot(x, y, label='Sine Wave')
plt.title('Matplotlib Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.show()
```

## 2. Seaborn
Seaborn builds on Matplotlib to provide beautiful statistical plots.

```python
# Import Seaborn
# Import Seaborn and Matplotlib
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Tips dataset
tips = sns.load_dataset('tips')

# Compute the correlation matrix using only numeric columns
numeric_tips = tips.select_dtypes(include=['float64', 'int64'])
correlation_matrix = numeric_tips.corr()

# Create a heatmap
plt.figure(figsize=(8, 5))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Seaborn Heatmap of Numeric Features in Tips Dataset')
plt.show()
```

## 4. Gapminder Dataset Visualization with Plotly
Using the Gapminder dataset, we create interactive scatter plots.

```python
import plotly.express as px
import pandas as pd

# Load the Gapminder-like dataset
gapminder = pd.read_csv('gapminder_dataset.csv')

# Create an interactive scatter plot
fig = px.scatter(
    gapminder,
    x='GDP per Capita',
    y='Life Expectancy',
    size='Population',  # Bubble size represents population
    color='Country',
    animation_frame='Year',  # Animate over years
    title='Interactive Gapminder-like Scatter Plot'
)
fig.show()
```

## 6. Airbnb Dataset Mapping with Folium
Using the Airbnb dataset, we map locations and prices.

```python
# Load the Airbnb dataset
airbnb = pd.read_csv('airbnb_dataset.csv')

# Create a Folium map
import folium
map_airbnb = folium.Map(location=[40.7, -73.85], zoom_start=12)

# Add markers for each Airbnb listing
for _, row in airbnb.iterrows():
    folium.Marker(
        location=[row['Latitude'], row['Longitude']],
        popup=f"Price: ${row['Price']} | Rating: {row['Rating']:.1f}"
    ).add_to(map_airbnb)

# Display the map
map_airbnb
```

## 7. Social Network Visualization with NetworkX
Using a sample social network dataset, we visualize connections.

```python
# Load the social network edges
social_edges = pd.read_csv('social_network_edges.csv')

# Create a graph with NetworkX
import networkx as nx
G = nx.from_pandas_edgelist(social_edges, source='source', target='target')

# Draw the graph
plt.figure(figsize=(8, 8))
nx.draw(G, with_labels=True, node_color='lightblue', edge_color='gray', node_size=500)
plt.title('Social Network Graph')
plt.show()
```
